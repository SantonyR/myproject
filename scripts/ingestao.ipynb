```python
import duckdb
import pandas as pd
import os
from datetime import datetime
```

```python
# cria uma conexão e salva os dados em um arquivo DuckDB
con = duckdb.connect(database='dados_duckdb.db', read_only=False)
```

# Camada Bronze

```python
# lê o arquivo CSV da landing zone
arquivo = 'z0019_2.csv'
data_ingestao = datetime.now()
df = pd.read_csv(f'../landing/{arquivo}', sep=';')
df['nome_arquivio'] = arquivo
df['data_ingestao'] = data_ingestao
df
```

```python
#con.execute('''drop table if exists bronze_z0019''')
```

```python
# criação da tabela bronze_z0019 no DuckDB
con.execute('''
    CREATE TABLE IF NOT EXISTS bronze_z0019 (
        NATBR VARCHAR,
        MAKTX VARCHAR,
        WERKS VARCHAR,
        MAINS VARCHAR,
        LABST VARCHAR,
        nome_arquivio VARCHAR,
        data_ingestao TIMESTAMP
    )
''')
```

```python
con.execute('INSERT INTO bronze_z0019 SELECT * FROM df')
resultado = con.execute('SELECT * FROM bronze_z0019').fetchdf()
resultado.head(10)
```
